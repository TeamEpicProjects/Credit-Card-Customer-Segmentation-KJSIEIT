### We found out that there are 4 types of purchase behaviour in the data set. So we need to derive a categorical variable based on their behaviour

```python
def purchase(df):
    if (df['ONEOFF_PURCHASES']==0) & (df['INSTALLMENTS_PURCHASES']==0):
        return 'none'
    if (df['ONEOFF_PURCHASES']>0) & (df['INSTALLMENTS_PURCHASES']>0):
         return 'both_oneoff_installment'
    if (df['ONEOFF_PURCHASES']>0) & (df['INSTALLMENTS_PURCHASES']==0):
        return 'one_off'
    if (df['ONEOFF_PURCHASES']==0) & (df['INSTALLMENTS_PURCHASES']>0):
        return 'installment'
```

```python
df['purchase_type']=df.apply(purchase,axis=1)
```

```python
df['purchase_type'].value_counts()
```

```python
df=pd.concat([df,pd.get_dummies(df['purchase_type'])],axis=1)
```

```python
df.head()
```

```python
df = df.drop(columns=['purchase_type'],axis=1)
```

## Data Normalization

```python
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df) 

# Normalizing the Data 
normalized_df = normalize(scaled_df) 
  
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df)
```

```python
scaled_df.shape
```

```python
scaled_df
```

```python
# Select the optimal number of clusters according to "Elbow Method"

scores_1 = []

range_values = range(1, 18)

for i in range_values:
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(scaled_df)
  scores_1.append(kmeans.inertia_) 

plt.plot(scores_1, 'bx-')
plt.title('Deciding the right number of clusters')
plt.xlabel('Clusters')
plt.ylabel('Scores') 
plt.show()
```

### k=6,8

```python
silhouette_scores = []

for n_cluster in range(2, 6):
    silhouette_scores.append( 
        silhouette_score(scaled_df, KMeans(n_clusters = n_cluster).fit_predict(scaled_df))) 
    
# Plotting a bar graph to compare the results 
k = [2, 3, 4, 5] 

plt.bar(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show()
```

## k=6

```python
# Apply K-means algorithms
kmeans = KMeans(6)
kmeans.fit(scaled_df)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
clusters=pd.concat([df, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head()
```

```python
# Principal component analysis
pca = PCA(n_components=2)
principal = pca.fit_transform(scaled_df)
principal
```

```python
# Create a new dataframe with the two components
pca_df = pd.DataFrame(data = principal, columns =['pca1','pca2'])
pca_df.head()
```

```python
# Concatenate the clusters labels to the dataframe
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})], axis = 1)
pca_df.head()
```

```python
plt.figure(figsize=(10,10))
ax = sns.scatterplot(x="pca1", y="pca2", hue = "cluster", data = pca_df, palette =['red','green','blue','pink','yellow','gray'])
plt.show()
```
